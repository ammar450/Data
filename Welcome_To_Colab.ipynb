```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
# Step 2: Import necessary libraries
import pyart
import matplotlib.pyplot as plt
import numpy as np
import glob
import os
import rasterio
from rasterio.transform import from_origin
from pyproj import Proj, transform
from scipy.interpolate import griddata

# Step 3: Define functions for polar to Cartesian conversion and saving as GeoTIFF
def polar_to_cartesian(radar, field, grid_size=1000, grid_extent=150.0):
    """
    Convert radar data from polar to Cartesian coordinates.

    Parameters:
    - radar: Py-ART Radar object
    - field: str, radar field to convert (e.g., 'reflectivity')
    - grid_size: int, number of grid points along each dimension
    - grid_extent: float, extent of the grid in km

    Returns:
    - cartesian_grid: 2D numpy array, the Cartesian grid of radar data
    - x: 1D numpy array, x-coordinates
    - y: 1D numpy array, y-coordinates
    """
    azimuths = radar.azimuth['data']
    ranges = radar.range['data'] / 1000.0  # Convert from meters to kilometers

    x = np.linspace(-grid_extent, grid_extent, grid_size)
    y = np.linspace(-grid_extent, grid_extent, grid_size)
    X, Y = np.meshgrid(x, y)

    R, Theta = np.meshgrid(ranges, np.deg2rad(azimuths))
    X_polar = R * np.sin(Theta)
    Y_polar = R * np.cos(Theta)

    points = np.vstack((X_polar.ravel(), Y_polar.ravel())).T
    values = radar.fields[field]['data'].ravel()

    cartesian_grid = griddata(points, values, (X, Y), method='linear')

    return cartesian_grid, x, y

def save_geotiff(output_path, data, x, y, crs="EPSG:4326"):
    """
    Save a Cartesian grid as a GeoTIFF file.

    Parameters:
    - output_path: str, path to save the GeoTIFF file
    - data: 2D numpy array, the data to save
    - x: 1D numpy array, x-coordinates
    - y: 1D numpy array, y-coordinates
    - crs: str, coordinate reference system (default is "EPSG:4326" for WGS 84)
    """
    transform = from_origin(x.min(), y.max(), (x[1] - x[0]), (y[1] - y[0]))

    with rasterio.open(
        output_path,
        'w',
        driver='GTiff',
        height=data.shape[0],
        width=data.shape[1],
        count=1,
        dtype=data.dtype,
        crs=crs,
        transform=transform
    ) as dst:
        dst.write(data, 1)
```

```
import os
import glob
import numpy as np
import pyart
from scipy.interpolate import griddata
import rasterio
from rasterio.transform import from_origin

# Step 1: Define a function to remove previous .tif files
def clean_tif_files(folder):
    tif_files = glob.glob(os.path.join(folder, '*.tif'))
    for tif_file in tif_files:
        print(f"Removing file: {tif_file}")
        os.remove(tif_file)

# Step 2: Define functions for polar to Cartesian conversion and saving as GeoTIFF
def polar_to_cartesian(radar, field, grid_size=1000, grid_extent=150.0):
    """
    Convert radar data from polar to Cartesian coordinates.

    Parameters:
    - radar: Py-ART Radar object
    - field: str, radar field to convert (e.g., 'reflectivity')
    - grid_size: int, number of grid points along each dimension
    - grid_extent: float, extent of the grid in km

    Returns:
    - cartesian_grid: 2D numpy array, the Cartesian grid of radar data
    - x: 1D numpy array, x-coordinates
    - y: 1D numpy array, y-coordinates
    """
    azimuths = radar.azimuth['data']
    ranges = radar.range['data'] / 1000.0  # Convert from meters to kilometers

    # Create a regular Cartesian grid
    x = np.linspace(-grid_extent, grid_extent, grid_size)
    y = np.linspace(-grid_extent, grid_extent, grid_size)
    X, Y = np.meshgrid(x, y)

    # Convert radar's polar coordinates to Cartesian
    R, Theta = np.meshgrid(ranges, np.deg2rad(azimuths))
    X_polar = R * np.sin(Theta)
    Y_polar = R * np.cos(Theta)

    # Interpolate the radar field to the Cartesian grid
    points = np.vstack((X_polar.ravel(), Y_polar.ravel())).T
    values = radar.fields[field]['data'].ravel()

    # Interpolation using griddata (can use 'linear', 'nearest', or 'cubic' methods)
    cartesian_grid = griddata(points, values, (X, Y), method='linear')

    return cartesian_grid, x, y

def save_geotiff(output_path, data, x, y, crs="EPSG:4326"):
    """
    Save a Cartesian grid as a GeoTIFF file.

    Parameters:
    - output_path: str, path to save the GeoTIFF file
    - data: 2D numpy array, the data to save
    - x: 1D numpy array, x-coordinates
    - y: 1D numpy array, y-coordinates
    - crs: str, coordinate reference system (default is "EPSG:4326" for WGS 84)
    """
    # Ensure the transform is set correctly (with georeferencing in mind)
    transform = from_origin(x.min(), y.max(), (x[1] - x[0]), (y[1] - y[0]))

    with rasterio.open(
        output_path,
        'w',
        driver='GTiff',
        height=data.shape[0],
        width=data.shape[1],
        count=1,
        dtype=data.dtype,
        crs=crs,  # Ensure CRS is properly assigned
        transform=transform
    ) as dst:
        dst.write(data, 1)

# Step 3: Define the path to radar files and clean the directory
data_folder = 'data/'  # Update this path if necessary

# Remove previous .tif files
clean_tif_files(data_folder)

# Step 4: Process each radar file and convert to GeoTIFF
files = glob.glob(data_folder + '*.nc')  # Only process .nc files

for file in files:
    print(f"\nProcessing file: {file}")
    try:
        radar = pyart.io.read(file)  # Ensure the file format is readable by Py-ART

        # Display radar metadata
        print("Radar information:")
        radar.info('compact')

        # Show the capture date of the radar file
        capture_date = radar.time['units']
        print(f"Capture date: {capture_date}")

        # Show the radar's name that captured the data
        radar_name = radar.metadata['instrument_name']
        print(f"Radar name: {radar_name}")

        # List available polarimetric variables
        available_variables = list(radar.fields)
        print("Available polarimetric variables:")
        print(available_variables)

        # Convert each variable from polar to Cartesian grid and save as GeoTIFF
        for var in available_variables:
            print(f"Processing variable: {var}")
            cartesian_grid, x, y = polar_to_cartesian(radar, var)
            output_path = file.split('.')[0] + f'_{var}_cartesian.tif'
            save_geotiff(output_path, cartesian_grid, x, y)
            print(f"Saved georeferenced GeoTIFF: {output_path}")
    except Exception as e:
        print(f"Error processing {file}: {e}")
```
